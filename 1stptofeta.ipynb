```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
```

```python
heart= pd.read_csv('etadata.csv')
```

```python
heart.head()


```

```python
# Function to add count labels to the bars
def add_count_labels(ax, data):
    for p in ax.patches:
        ax.annotate(f'\n{int(p.get_width())}', (p.get_width(), p.get_y()), ha='left', va='center', color='black', fontsize=10)


# GENHEALTH COUNT
print("Number of Categories :", len(heart["GenHealth"].unique()))
print(heart["GenHealth"].value_counts())
plt.figure(figsize=(10, 6))
ax = sb.countplot(y="GenHealth", data=heart, order=heart["GenHealth"].value_counts().index)
add_count_labels(ax, heart["GenHealth"].value_counts())
plt.title('Count of GenHealth')
plt.xlabel('Count')
plt.ylabel('GenHealth')
plt.show()

# DIFFWALKING COUNT
print("Number of Categories :", len(heart["DiffWalking"].unique()))
print(heart["DiffWalking"].value_counts())
plt.figure(figsize=(10, 6))
ax = sb.countplot(y="DiffWalking", data=heart, order=heart["DiffWalking"].value_counts().index)
add_count_labels(ax, heart["DiffWalking"].value_counts())
plt.title('Count of DiffWalking')
plt.xlabel('Count')
plt.ylabel('DiffWalking')
plt.show()

# AGECATEGORY COUNT
print("Number of Categories :", len(heart["AgeCategory"].unique()))
print(heart["AgeCategory"].value_counts())
plt.figure(figsize=(10, 6))
ax = sb.countplot(y="AgeCategory", data=heart, order=heart["AgeCategory"].value_counts().index)
add_count_labels(ax, heart["AgeCategory"].value_counts())
plt.title('Count of AgeCategory')
plt.xlabel('Count')
plt.ylabel('AgeCategory')
plt.show()

# PHYSICAL HEALTH COUNT
print("Number of Categories :", len(heart["PhysicalHealth"].unique()))
print(heart["PhysicalHealth"].value_counts())
plt.figure(figsize=(10, 6))
ax = sb.countplot(y="PhysicalHealth", data=heart, order=heart["PhysicalHealth"].value_counts().index)
add_count_labels(ax, heart["PhysicalHealth"].value_counts())
plt.title('Count of PhysicalHealth')
plt.xlabel('Count')
plt.ylabel('PhysicalHealth')
plt.show()

# STROKE COUNT
print("Number of Categories :", len(heart["Stroke"].unique()))
print(heart["Stroke"].value_counts())
plt.figure(figsize=(10, 6))
ax = sb.countplot(y="Stroke", data=heart, order=heart["Stroke"].value_counts().index)
add_count_labels(ax, heart["Stroke"].value_counts())
plt.title('Count of Stroke')
plt.xlabel('Count')
plt.ylabel('Stroke')
plt.show()
```
