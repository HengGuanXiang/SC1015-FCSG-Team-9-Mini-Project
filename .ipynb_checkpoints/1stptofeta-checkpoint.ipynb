```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
```

```python
heart= pd.read_csv('selected_heartdata.csv')
```

```python
heart.head()


```

```python
# GENHEALTH COUNT
print("Number of Categories :", len(heart["GenHealth"].unique()))
print(heart["GenHealth"].value_counts())
sb.catplot(y = "GenHealth", data = heart, kind = "count", height = 8)

# DIFFWALKING COUNT
print("Number of Categories :", len(heart["DiffWalking"].unique()))
# Data of each DiffWalking
print(heart["DiffWalking"].value_counts())
sb.catplot(y = "DiffWalking", data = heart, kind = "count", height = 8)

# STROKE COUNT
print("Number of Categories :", len(heart["AgeCategory"].unique()))
print(heart["AgeCategory"].value_counts())
sb.catplot(y = "AgeCategory", data = heart, kind = "count", height = 8)

# PHYSICAL HEALTH COUNT
print("Number of Categories :", len(heart["PhysicalHealth"].unique()))
print(heart["PhysicalHealth"].value_counts())
sb.catplot(y = "PhysicalHealth", data = heart, kind = "count", height = 8)
```

```python


print(heart.dtypes)
PhysicalHealth = pd.DataFrame(heart['PhysicalHealth'])
```

```python
PhysicalHealth = pd.DataFrame(heart['PhysicalHealth'])
GenHealth = pd.DataFrame(heart['GenHealth'])
DiffWalking = pd.DataFrame(heart['DiffWalking'])
AgeCategory  = pd.DataFrame(heart['AgeCategory'])
HeartDisease = pd.DataFrame(heart['HeartDisease'])
```

```python


f = plt.figure(figsize=(16, 8))
sb.boxplot(data = PhysicalHealth, orient = "h")



f = plt.figure(figsize=(16, 8))
sb.boxplot(data = GenHealth, orient = "h")



f = plt.figure(figsize=(16, 8))
sb.boxplot(data = DiffWalking, orient = "h")


f = plt.figure(figsize=(16, 8))
sb.boxplot(data = AgeCategory, orient = "h")


```
